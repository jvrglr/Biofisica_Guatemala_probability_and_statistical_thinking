## Numerical activity: Simulation of betting games

In this activity we will numerically explore games of chance and test theoretical probability predictions using random number generation. We have studied two classical problems — Gombaud’s problem and the Monty Hall paradox — where there exists an optimal strategy that maximizes the probability of winning. However, a player who ignores probability theory might still occasionally win using a suboptimal choice due to randomness.

We are interested in understanding how the law of large numbers manifests in practice: if two players use different strategies, how many rounds must they play before the better strategy is overwhelmingly more successful?

For both the Gombaud and Monty Hall problems:

1. Simulate \(M\) independent plays of the game.  
2. Record the number of wins obtained under each strategy.  
3. With the data generated in 2, compute the empirical probability of winning and its empirical standard deviation.

Determine the minimum number of games \(M\) such that the probability that the optimal strategy leads to more wins exceeds \(90\%\).

Make a plot of the empirical probability of winning (with errorbars!) vs the number of games \(M\), and compare:

- The theoretical winning probability  
- The simulated winning probability obtained from the numerical runs  

This activity illustrates how statistical regularities emerge from repeated trials, and how empirical simulation converges toward theoretical probability.



```python
#Import libraries
import numpy as np #Do math
import matplotlib.pyplot as plt #Plotting
```

```python
#Example: evaluating the probability of winning a dice game
Ms = [20,100,1000]
p_wins = []
errs = []
for M in Ms:
    dice_rolls = np.random.randint(1, 7, size=M)
    p_win = np.sum(dice_rolls == 6) / M #Say that we win if we roll a 6
    std = np.sqrt(p_win * (1 - p_win) / M)
    err = 1.96 * std  # 95% confidence interval
    p_wins.append(p_win)
    errs.append(err)

plt.scatter(Ms, p_wins, marker='o', label='Estimated Probability of Winning')
plt.errorbar(Ms, p_wins, yerr=errs, fmt='o', capsize=5)
plt.axhline(y=1/6, color='r', linestyle='--', label='Theoretical Probability (1/6)')
plt.legend()
plt.xlabel('Number of Simulations (M)')
plt.ylabel('Probability of Winning')
```
